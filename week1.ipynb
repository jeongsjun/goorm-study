```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm.auto import tqdm
import lxml
```

```python
URL='http://boannews.com/media/t_list.asp?Page=1&kind='
site=requests.get(URL)
site
```

```python
site.text
```

```python
site_soup=BeautifulSoup(site.text, 'lxml')
site_soup
```

```python
news_list=site_soup.select('div.news_list')
news_list
```

```python
data=pd.DataFrame(columns=['title','url','writer'])
data
```

```python


for news in tqdm(news_list):
    news_title=news.select('span.news_txt')[0].text
    news_url=news.find('a').get('href')
    news_url='http://boannews.com'+news_url
    news_writer=news.select('span.news_writer')[0].text

    new_data=[news_title, news_url, news_writer]
    data.loc[len(data)]=new_data

data
```

```python
data.to_csv('boannews.csv', encoding='utf-8-sig', index=False)
```
